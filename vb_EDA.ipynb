## Feature Engineering: Hitting Efficiency

To get a better sense of an attacker's true impact, I am calculating **Hitting Percentage**. Unlike raw kill counts, this metric penalizes errors and provides a normalized view of efficiency.

The calculation is performed using the following formula:

$$
\text{Hitting Percentage} = \frac{\text{Kills} - \text{Attacking Errors}}{\text{Kills} + \text{Attacking Errors} + \text{Attacking Attempts}}
$$

### Why this matters:
* **Efficiency over Volume:** A player with many kills might also have many errors, resulting in a low hitting percentage that hurts the team.
* **Normalization:** This allows us to compare players with different total attacking opportunities on a level playing field.

```python
df['Hitting Percentage'] = (df['Kills'] - df['Attacking Errors']) / (df['Kills'] + df['Attacking Errors'] + df['Attacking Attempts'])
```

## Team-Level Efficiency: Avoiding the "Average of Averages"

When transitioning from player-level data to team-level insights, the method of calculation is vital. To accurately represent a team's offensive efficiency, we must sum the raw statistics (Kills, Errors, and Attempts) for the entire roster before calculating the percentage.

### The Importance of Weighted Totals
If we simply averaged the individual hitting percentages of every player on the team, we would encounter the **"Average of Averages" fallacy**. This would lead to several issues:
* **Equal Weighting:** A bench player with only 2 attempts would have the same statistical influence on the team average as a starting Opposite with 100+ attempts.
* **Skewed Data:** Inactive or low-volume players could artificially inflate or deflate the team's true performance metric.

By using the sum of all actions first, the resulting **Team Hitting Percentage** correctly weights every contribution based on actual volume, providing a true reflection of the team's offensive output.

```python
team_hitting_percentage = (df.groupby('Team')['Kills'].sum() - df.groupby('Team')['Attacking Errors'].sum()) / (df.groupby('Team')['Kills'].sum() + df.groupby('Team')['Attacking Errors'].sum() + df.groupby('Team')['Attacking Attempts'].sum())
team_hitting_percentage.sort_values(ascending=False)
```

```python
# Team Hitting Percentage 
team_hitting_percentage.sort_values(ascending=False).plot(kind='bar')
plt.title('Team Hitting %')
plt.xlabel('Team')
plt.ylabel('Hitting %')
plt.show()
```

## Calculating Service Error Rate (Feature Engineering)

I am establishing a standardized **Service Error Rate** to evaluate serving consistency across the roster. This allows us to compare players based on their accuracy rather than just the raw number of mistakes.

The metric is calculated as follows:

$$
\text{Service Error Rate} = \frac{\text{Service Errors}}{\text{Service Attempts}}
$$

### Data Integrity Note:
In cases where a player has **0 Service Attempts**, this calculation would naturally result in an undefined (infinity) value. To ensure the dataset remains usable for visualizations and further analysis, I will follow this calculation with a step to replace any resulting `NaN` or `inf` values with `0`. This reflects the fact that a player who hasn't served cannot have an error rate.

```python
df['Service Error rate'] = df['Service Errors'] / df['Service Attempts']
```

## Service Strategy and Team Style

This chart reveals which teams play it safe and which teams take big risks at the service line. A team's service error rate is a great window into their overall personality and tactical approach.

### Topspin vs. Float Serves
The disparity in error rates usually comes down to the type of serve being utilized:
* **High-Risk Serving:** Teams with higher error rates typically favor hard topspin serves. These are harder to pass but much more difficult to keep in the court.
* **Low-Risk Serving:** Teams with lower error rates often focus on float serves or controlled topspin. These are safer and more likely to result in a rally rather than a direct point or error.

### Calculating Team Performance
To get a true reading of a team's efficiency, we calculate the error rate using the total sums of the roster. This avoids the mistake of treating a player with one serve the same as a player with one hundred.

$$\text{Service Error Rate} = \frac{\sum \text{Team Service Errors}}{\sum \text{Team Service Attempts}}$$

Analyzing these numbers helps identify which teams are comfortable with high-variance play and which ones prefer a more stable, defensive game.

```python
error_rate_by_age = df.groupby('Team')['Service Errors'].sum() / df.groupby('Team')['Service Attempts'].sum()
error_rate_by_age = error_rate_by_age.sort_values(ascending=False)
error_rate_by_age.plot(kind='bar')
plt.title('Service Error rate by Team')
plt.xlabel('Team')
plt.ylabel('Service Error Rate')
plt.show()
```

## Data Validation and Filtering

In this step, I am filtering the dataset to ensure we are only analyzing logically sound records. The specific condition used is:

`serve = df[(df['Service Errors'] < df['Service Attempts']) & (df['Service Attempts] > 0)].copy()`

### Why this filter is necessary:
* **Integrity Check:** Mathematically, a player cannot commit more service errors than the total number of times they attempted to serve. 
* **Preventing Skewed Data:** Including rows where errors exceed attempts would result in an error rate greater than 100%. This is often the result of manual data entry mistakes or "ghost" statistics that can skew the team averages and lead to inaccurate visualizations.
* **Cleaning for Analysis:** By removing these anomalies, we ensure that our ranking of the top 10 servers is based on realistic, high-quality data.

The resulting table displays the most efficient servers based on **Aces**, while keeping the data grounded in realistic attempt-to-error ratios.

```python
serve = df[(df['Service Errors'] < df['Service Attempts']) & (df['Service Attempts'] > 0)].copy()
serve.loc[:, ['Player Name', 'Service Error rate', 'Service Attempts', 'Service Errors', 'Aces']].sort_values(by='Aces', ascending=False).head(10)
```
