Correlation

```python
numeric_cols = df.select_dtypes(include=['int64', 'float32']).columns
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

## Correlation Matrix Analysis

The correlation matrix reveals strong, logical relationships between several key performance variables.

### Offensive Dynamics
We observe that **kills** are highly correlated with **aces, service errors, service attempts, attacking errors, and attacking attempts**. 
* **Volume:** More kills naturally require more attacking attempts.
* **Risk/Reward:** Increased attacking attempts lead to more attacking errors. 
* **Aggression:** Strong offensive players tend to use aggressive topspin serves rather than safe float serves. This strategy increases their service attempts and aces but also results in more service errors.

### Defensive Dynamics
The data also shows a logical connection between defensive metrics:
* **Recovery:** **Defensive errors** are highly correlated with **great saves**. This makes sense because a great save is usually required only after a ball has been misplayed into a difficult position.
* **Execution:** **Defensive errors** and **defensive receptions** are linked because most errors in this category stem from mistakes made during the act of receiving or playing the ball defensively.

### This validates some of the statistics in the correlation matrix

Team Hitting Percentage

```python
"""Because this dataset is made up of players, each with their own number of kills, errors, and attempts. We must sum the values for each stat per team
before we take any average, this avoids taking what is known as an "average of averages". If this was not done then bench players would be weighted equally
as players who played most of the game and could inflate or deflate the stats of a team."""
team_hitting_percentage = (df.groupby('Team')['Kills'].sum() - df.groupby('Team')['Attacking Errors'].sum()) / (df.groupby('Team')['Kills'].sum() + df.groupby('Team')['Attacking Errors'].sum() + df.groupby('Team')['Attacking Attempts'].sum())
team_hitting_percentage.sort_values(ascending=False)
```

```python
team_hitting_percentage.sort_values(ascending=False).plot(kind='bar')
plt.title('Team Hitting %')
plt.xlabel('Team')
plt.ylabel('Hitting %')
plt.show()
```

```python
age_aces_df = df.groupby('Age')['Aces'].sum()
plt.plot(age_aces_df)
```

```python
df['Service Error rate'] = df['Service Errors'] / df['Service Attempts']
```

```python
"""From this chart we can see which teams are more conservative(serving with less aggression, focused on keeping it in) and teams
who are less conservative(serving with more aggression, focused on acing). This is heavily correlated with the teams dynamics and 
says a lot about their style of play and personality. Error rate is derived from the sum of a teams service errors divided by the 
sum of their service attempts. Teams with higher error rate tend to serve more topspin serves, these are more likely to serve out 
or go into the net but have a higher chance of acing, teams with lower error rate tend to serve more float serves or lighter topspin 
serves, these are more likely to go in but have a lower chance of acing. 
"""
error_rate_by_age = df.groupby('Team')['Service Errors'].sum() / df.groupby('Team')['Service Attempts'].sum()
error_rate_by_age = error_rate_by_age.sort_values(ascending=False)
error_rate_by_age.plot(kind='bar')
plt.title('Error rate by Team')
plt.xlabel('Team')
plt.ylabel('Error Rate')
plt.show()
```
